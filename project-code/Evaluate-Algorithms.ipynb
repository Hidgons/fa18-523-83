```python
def split_data(df):
    x = df.iloc[:,2:-1]
    y = df.iloc[:,1]
    return x, y
```

```python
x,y=split_data(df)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
def LogisticRegression_Penalized():
    algorithm= LogisticRegression(C=500,penalty='l1', tol=0.10, solver='saga')
    return algorithm
```

```python
algorithm=LogisticRegression()
```

```python
def model_training(x_train,y_train,algorithm):
    model=algorithm.fit(x_train, y_train)
    return model
```

```python
def model_prediction(x_test, model):
    y_pred = model.predict(x_test)
    return y_pred
```

```python

def score(algorithm,x,y):
    scoring = {'accuracy', 'roc_auc','average_precision','recall','precision','f1'}
    scores = cross_validate(algorithm, x, y, cv=10,    scoring=scoring ,return_train_score=True)
    return scores
```

```python
df_scores=pd.DataFrame(score(algorithm,x,y))
```
