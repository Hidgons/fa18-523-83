```python

def read_pickle(input_file):
    df = pd.read_pickle(input_file)
    return df
```

Define function to simplify column names

```python
def simple_label(df):
	df.columns=['id','y','x1','x2','x3','x4','x5']
	return df
```

read file from pickle generated by Makefile `make preprocessing-files` command

```python
df=read_pickle('../data/processed/training.pkl')
df
```

Define function to split dataset into y label and x variables

```python
def split_data(df):
    x = df.iloc[:,2:-1]
    y = df.iloc[:,1]
    return x, y
```

```python
x,y=split_data(df)
```

define function to customized LR Penalize Algorithm

```python
def LogisticRegression_Penalized():
    algorithm= LogisticRegression(C=500,penalty='l1', tol=0.10, solver='saga')
    return algorithm
```

define function to cross validate, split data , and output scorings

```python

def score(algorithm,x,y):
    scoring = {'accuracy', 'roc_auc','average_precision','recall','precision','f1'}
    scores = cross_validate(algorithm, x, y, cv=10,    scoring=scoring ,return_train_score=True)
    return scores
```

```python
df_score_all=pd.DataFrame()
df_score_all['Logistic Regression Penalized']=pd.DataFrame(score(LogisticRegression_Penalized(),x,y)).mean()
df_score_all['Random Forest']=pd.DataFrame(score(RandomForestClassifier(),x,y)).mean()
```

```python
def get_balanced_data(df):
    df_majority = df[df['y']==0]
    df_minority = df[df['y']==1]
    
    df_majority_downsampled = df_majority.sample( n=10026, random_state=123)
    df_downsampled = pd.concat([df_majority_downsampled, df_minority])
    return df_downsampled
```
